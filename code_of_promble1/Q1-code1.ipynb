```python
# 数据分析：项目类型对国家奖牌数量的影响
'''
通过数据预处理提取项目类型及其对应的比赛数量，同时聚合各国的奖牌数量数据。
通过比例分配的方法，将项目比赛数量与国家奖牌数进行关联，计算每种项目类型对奖牌分布的影响，并评估其相关性。
最后展示项目比赛数量在奖牌分布中的作用，从而为进一步的多变量建模提供数据支持和理论依据。
'''
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 数据预处理
project_columns = excel_data.iloc[:, 1:-5].columns  # 提取项目列
project_totals = excel_data[project_columns].select_dtypes(include=[float, int]).sum()  # 统计项目比赛数量

# 按国家聚合奖牌数据
country_medals = csv_data[['NOC', 'Gold', 'Silver', 'Bronze', 'Total']].groupby('NOC').sum()

# 构建数据框
project_medals = pd.DataFrame({
    'Project': project_columns[:len(project_totals)],  # 确保列名长度匹配
    'Event_Count': project_totals.values
})

# 计算每个项目类型对奖牌数量的贡献
##首先，统计每个项目的比赛场次占总比赛的比例，作为项目的权重；统计每个国家的奖牌数量，并计算这些奖牌占该项目总奖牌的比例；
#接着，将项目权重与国家在该项目中的奖牌比例相乘，得到每个项目对该国家奖牌的贡献；
#最后，将所有项目的贡献值加总，得出各项目对各国奖牌数量的整体影响。
#需要注意 这里不能单个国家的来看 因为并每个国家的在各自项目上的奖牌数量 只能看汇总值。
correlations = []
for medal_type in ['Gold', 'Silver', 'Bronze', 'Total']:
    project_medals[medal_type] = [
        country_medals[medal_type].sum() * count / project_totals.sum()  # 按比例分配奖牌
        for count in project_medals['Event_Count']
    ]
    correlations.append((medal_type, project_medals[medal_type].corr(project_medals['Event_Count'])))
```

```python
# 对项目数量降序排序
project_medals_sorted = project_medals.sort_values(by='Event_Count', ascending=False)

# 可视化：降序排序后的柱状图
plt.figure(figsize=(10, 18))
sns.barplot(
    x=project_medals_sorted['Event_Count'],
    y=project_medals_sorted['Project'],
    palette="viridis"
)
plt.title("Effects of Competition Counts on Medal Numbers")  # 英文标题
plt.xlabel("Number of Projects")  # X轴标签
plt.ylabel("Project Type")  # Y轴标签
plt.show()
```
