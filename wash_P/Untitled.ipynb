```python
import pandas as pd
import os

# 读取 CSV 文件
file_path = "summerOly_athletes.csv"  # 替换为你的 CSV 文件路径
df = pd.read_csv(file_path)

# 定义列名
columns = ["NOC", "Year", "Event"]  # 替换为你的列名

# 检查列名是否有效
if all(col in df.columns for col in columns):
    # 遍历每一行
    for i in range(len(df)):
        # 提取第 i 行的指定列的值
        row_values = df.loc[i, columns].tolist()  # 转换为列表
        print(f"第 {i} 行的 {columns} 列的值:", row_values)

        # 获取 NOC 名称
        noc_name = row_values[0]  # NOC 列的值
        year = row_values[1]  # Year 列的值
        event = row_values[2]  # Event 列的值

        # 构造对应的 Excel 文件名
        excel_file = f"{noc_name}.xlsx"

        # 检查 Excel 文件是否存在
        if os.path.exists(excel_file):
            # 读取 Excel 文件
            df_excel = pd.read_excel(excel_file, index_col=0)  # 假设第一列是行名

            # 检查 Year 行和 Event 列是否存在
            if year in df_excel.index and event in df_excel.columns:
                # 对指定位置的数据加 1
                df_excel.loc[year, event] += 1

                # 保存修改后的 Excel 文件
                df_excel.to_excel(excel_file)
                print(f"文件 {excel_file} 的 {year} 行 {event} 列数据已加 1。")
            else:
                print(f"文件 {excel_file} 中不存在 {year} 行或 {event} 列。")
        else:
            print(f"文件 {excel_file} 不存在。")
else:
    print("列名无效，请检查输入。")
```
